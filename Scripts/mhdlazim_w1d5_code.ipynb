Task :- 2
Data health inspection

```python
print("shape:",df.shape)
print()
print(df.dtypes)
print()
print()
print(df.isnull().sum())
print()
print(df[["Sales","Quantity","Discount","Profit"]].describe())
```

Task :- 3
Check and remove duplicates

```python
print("Duplicate Rows:",df.duplicated().sum())
dupes = df[df.duplicated(subset=["Order ID"], keep=False)]
print(dupes.shape)
dupes.head()
```

Task :- 4
Convert date columns

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
print(df.dtypes)
```

Task :- 5
Engineer Shipping Days

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
print(df.groupby("Region")["Shipping Days"].mean().round(1))
```

Task :- 7
Engineer Profit Margin %

```python
df["Profit Margin %"] = (df["Profit"] / df["Sales"] * 100).round(2)
print(df.groupby("Sub-Category")["Profit Margin %"].mean()
      .sort_values().round(2))
```

Task:-6 Engineer Order Year

```python
df["Order year"] = df["Order Date"].dt.year
```
